```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://engineering.careers360.com/colleges/ranking/top-engineering-colleges-in-maharashtra?sort_by=1&degree=2&exam=5755&ownership=2%2C1&branch=70%2C181&stream=1&state=15',headers=headers).text
```

```python
soup=BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
college = soup.find_all('div', class_='tupple')
```

```python
len(college)
```

```python
name = []
rating = []
reviews = []
cut_off = []
placement = []
Coureses = []
ownership = []
for i in college:
    
    name.append(i.find('h3').text.strip())
    rating.append(i.find('b'))
    reviews.append(i.find('span', class_='review_text').text.strip())
    ownership.append(i.find('strong',class_='strong_ownership').text.strip())
    cut_off.append(i.find('ul', class_='links_list').text.strip())
```
