```python
# Simple linear Regression
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('house_prices.csv')
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.scatter(dataset['Bedrooms'], dataset['Price'], marker='o')
plt.title('Bedrooms vs Prices')
plt.xlabel('Bedrooms available')
plt.ylabel('Prices available')
plt.grid(True)  # Add a grid if desired
plt.show()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
```

```python
df.head()
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('House vs Prices (Training set)')
plt.xlabel('House')
plt.ylabel('Prices')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('House vs Prices (Test set)')
plt.xlabel('House')
plt.ylabel('Prices')
plt.show()
```
