```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
```

```python

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Get user input for the number of clusters (k)
k = int(input("Enter the number of clusters (k): "))
```

```python
# Create k-means model
kmeans = KMeans(n_clusters=k, random_state=42)
```

```python
# Fit the model to the scaled data
kmeans.fit(X_scaled)
```

```python
# Get cluster assignments and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Visualize the results (for simplicity, consider only the first two features)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', marker='o', s=50, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title(f'K-Means Clustering (k={k}) - Iris Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.legend()
plt.show()
```
